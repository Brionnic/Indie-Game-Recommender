```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import rc
rc('text', usetex=True)
%matplotlib inline
```

```python
x_list = np.arange(1,60000)
z_list = np.zeros(59999)
```

```python
print x_list
print
print z_list
```

```python
y_list = []

for x in x_list:
    if x < 20:
        temp = (x/20.0)**2
        y_list.append(temp)
    else:
        y_list.append(1.0)
```

```python
print y_list
```

```python
fig, ax = plt.subplots(figsize=(12,4))

ax.plot(x_list, y_list, color="#2CE22C", lw=3)
ax.fill_between(x_list, z_list, y_list, color="#2CD22C", alpha=0.5)
ax.set_xscale("log")
ax.set_ylim(0,1.025)
ax.set_xlim(0, 60000)
ax.set_facecolor("#272D27")
# ax.set_yticklabels
ax.set_yticklabels(["0%", "20%", "40%", "60%", "80%", "100%"], size=16)
ax.set_xticklabels(["0", "1", "10", "100", "1000", "10000"], size=16)
ax.set_xlabel("Number of Users (Log Scale)", size=20)
ax.set_ylabel("% of Rating Remaining", size=18)
ax.set_title(r"Low Unique Users Weighting Applied\nU<20 $R_w = R \times(\frac{U}{20})^2$", size=22)
```

```python
xx_list = np.arange(1,60000)
zz_list = np.zeros(59999)

yy_list = []

for x in xx_list:
    if x < 20:
        temp = (x / 20.0)**2
        yy_list.append(temp)
    elif x >  10000:
        temp = (10000/(x*1.0))
        yy_list.append(temp)
    else:
        yy_list.append(1.0)
```

```python
fig, ax1 = plt.subplots(figsize=(12,4))

ax1.plot(xx_list, yy_list, color="#2CE22C", lw=3)
ax1.fill_between(xx_list, zz_list, yy_list, color="#2CD22C", alpha=0.5)
ax1.set_xscale("log")
ax1.set_ylim(0,1.025)
ax1.set_xlim(0, 60000)
ax1.set_facecolor("#272D27")
ax1.set_yticklabels(["0%", "20%", "40%", "60%", "80%", "100%"], size=16)
ax1.set_xticklabels(["0", "1", "10", "100", "1000", "10000"], size=16)
ax1.set_xlabel("Number of Users (Log Scale)", size=20)
ax1.set_ylabel("% of Rating Remaining", size=18)
ax1.set_title("Stage 2 Weighting Applied", size=22)
```

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
%matplotlib inline


data = pd.read_csv("rmse.csv")
```

```python
data.columns = ["rmse_train", "rmse_test", "rank"]
```

```python
x_line = np.arange(1,31)

avg_line = np.array([1.77 for x in x_line])

fig, bx = plt.subplots(figsize=(12,6))
bx.plot(x_line, avg_line, color="green", alpha=0.8, lw=3, linestyle="--",label="Average as Prediction")
bx.plot(data["rank"], data["rmse_train"], color="#2CE22C",linestyle="dotted",  alpha=0.7, lw=3, label="RMSE - Train")
bx.plot(data["rank"], data["rmse_test"], color="#2CE22C", alpha=0.9, lw=3, label="RMSE - Test")
bx.set_xlim(1,30)
bx.set_ylim(1.6, 1.85)
bx.set_xlabel("Rank Hyperparameter", size=22)
bx.set_ylabel("RSME", size=22)
bx.set_title("RMSE vs Rank", size=24)
bx.set_xticklabels(["1","5","10","15","20","25","30"], size=16)
bx.set_yticklabels(["1.60", "1.65", "1.70", "1.75", "1.80", "1.85"], size=16)
bx.legend()
bx.set_facecolor("#272D27")
```
