```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
%matplotlib inline
```

```python
model_0 = pd.read_csv("nan_to_mean_rmse_lpm_b0_s0.csv")
model_0.columns=["rmse_train", "rmse_test", "rank"]
print model_0.head()
model_0.keys()
```

```python
fig, axmodel0 = plt.subplots(figsize=(12,6))

axmodel0.plot(model_0["rank"], model_0["rmse_train"], color="green", alpha=0.4, label="RMSE Train")
axmodel0.plot(model_0["rank"], model_0["rmse_test"], color="green", alpha=0.8, label="RMSE Test")
axmodel0.set_xlim(1,10)
axmodel0.set_ylim(.7, 1)

_x = np.arange(1,11)
_y = np.array([0.922, 0.922, 0.922, 0.922, 0.922, 0.922, 0.922, 0.922, 0.922, 0.922])

axmodel0.plot(_x, _y, color="grey", alpha=0.8, label="RMSE from Average")


axmodel0.set_ylabel("RMSE")
axmodel0.set_xlabel("Hyperparameter: Rank")

axmodel0.legend()
```

```python
model_1 = pd.read_csv("nan_to_mean_rmse_lpm_b0_s1.csv")
model_1.columns=["rmse_train", "rmse_test", "rank"]
print model_1.head()
model_1.keys()

fig, axmodel1 = plt.subplots(figsize=(12,6))

axmodel1.plot(model_1["rank"], model_1["rmse_train"], color="green", alpha=0.4, label="RMSE Train", lw=3)
axmodel1.plot(model_1["rank"], model_1["rmse_test"], color="green", alpha=0.8, label="RMSE Test", lw=3)
axmodel1.set_xlim(1,10)
axmodel1.set_ylim(.7, 1)

_x = np.arange(1,11)
_y = np.array([0.921, 0.921, 0.921, 0.921, 0.921, 0.921, 0.921, 0.921, 0.921, 0.921])

axmodel1.plot(_x, _y, color="grey", alpha=0.8, label="RMSE Test")

axmodel1.set_ylabel("RMSE", size=14)
axmodel1.set_xlabel("Hyperparameter: Rank", size=14)

axmodel1.set_title("RMSE vs Rank Hyperparameter\n(Log Minutes Played, Weighting Stage 1)", size=18)

axmodel1.legend()
```
